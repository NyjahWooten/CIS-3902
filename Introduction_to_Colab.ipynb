```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline
sns.set(color_codes=True)
```

```python
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
df.head(5)
```

```python
df = df.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price" })
df.head(5)
```

```python
numerical_df = df.select_dtypes(include=np.number)  # Select numerical columns
Q1 = numerical_df.quantile(0.25) # Calculate the 25th percentile
Q3 = numerical_df.quantile(0.75) # Calculate the 75th percentile
IQR = Q3 - Q1
print(IQR)
```

```python
# Select numerical columns
numerical_df = df.select_dtypes(include=np.number)
# Calculate quantiles and IQR
Q1 = numerical_df.quantile(0.25)
Q3 = numerical_df.quantile(0.75)
IQR = Q3 - Q1

# Align the DataFrame and Series using their indexes
# (This step is crucial to fix the alignment issue)
df, Q1 = df.align(Q1, axis=1, copy=False)
df, Q3 = df.align(Q3, axis=1, copy=False)
df, IQR = df.align(IQR, axis=1, copy=False)

# Now perform the outlier removal
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

```python
df.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

```python
plt.figure(figsize=(10,5))
# Calculate the correlation matrix for numerical columns only
c = df.select_dtypes(include=np.number).corr()
sns.heatmap(c, cmap="BrBG", annot=True)
c
```
